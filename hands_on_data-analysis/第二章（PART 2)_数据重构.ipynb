### 数据重构

```python
import pandas as pd
import numpy as np
text_left_up = pd.read_csv('data_0/train-left-up.csv')
text_left_down = pd.read_csv('data_0/train-left-down.csv')
text_right_up = pd.read_csv('data_0/train-right-up.csv')
text_right_down = pd.read_csv('data_0/train-right-down.csv')
```

```python
text_left_up.head()
```

```python
text_left_down.head()
```

```python
text_right_down.head()
```

```python
text_right_up.head()
```

```python
list_up = [text_left_up,text_right_up]
result_up = pd.concat(list_up,axis = 1)
result_up.head()
```

```python
list_down = [text_right_down,text_left_down]
result_down = pd.concat(list_down,axis = 1)
result_down.head()
```

```python
result = pd.concat([result_up,result_down])
result.head()
```

### 使用DataFrame自带的方法join和append

```python
result_up = text_left_up.join(text_right_up)
```

```python
result_down = text_left_down.join(text_right_down)
result = result_up.append(result_down)
```

```python
result.head()
```

### pandas 的merge 方法

```python
result_up = pd.merge(text_left_up,text_right_up,left_index=True,right_index=True)
result_down = pd.merge(text_left_down,text_right_down,left_index=True,right_index=True)

result = result_up.append(result_down)
result.head()
```

```python
result.to_csv('result.csv')
```
